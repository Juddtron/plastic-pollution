```python
import pandas as pd
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Resources/share_plastic_fate.csv")
```

```python
df.head()
```

```python
new_column_names = {
    'Share of waste recycled from total regional waste': 'Recycled',
    'Share of waste incinerated from total regional waste': 'Incinerated',
    'Share of littered and mismanaged from total regional waste': 'Mismanaged',
    'Share of waste landfilled from total regional waste': 'Landfilled'
}

df = df.rename(columns=new_column_names)
```

```python
df.head()
```

```python
new_df = df.drop(columns="Code")
```

```python
new_df
```

```python
column_to_round = ['Recycled', 'Incinerated', 'Mismanaged', 'Landfilled']
for column in column_to_round:
    new_df[column] = new_df[column].round(2)
new_df
```

```python
grouped = new_df.groupby(['Year', 'Entity'])
```

```python
grouped_mean = grouped.mean()
```

```python
grouped_mean.head(10)
```

```python
new_df.to_json('recycled_plastic.json', orient='records', indent=4)
```
